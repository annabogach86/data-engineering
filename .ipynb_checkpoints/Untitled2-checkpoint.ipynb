```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://belleyou.ru/catalog/odezhda/bryuki/'
```

```python
# получение ответа от сайта
response = requests.get(url)
response
```

```python
response.text
```

```python
# подключаем анализатор html кода lxml, что бы легко можно было находить элементы
soup = BeautifulSoup(response.text, 'lxml') # html.parser альтернатива
```

```python
soup
```

```python
# find вернет хэш_тег div 
data = soup.find("div", class_="product-list1__item catalog-item")
data
```

```python
name = data.find('a', class_='product-list1__title').text.replace('\n', '')
name
```

```python
price = data.find('span', class_='product-list1__price').text.strip()
f_price = price.replace('9\xa0', '')
f_price
```
